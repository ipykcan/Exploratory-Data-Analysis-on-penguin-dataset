```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('penguins_size.csv')
df
```

```python
print(df.shape)
```

```python
df.info()
```

```python
#removing unwanted data
df.dropna(inplace = True)
df
```

```python
#statistical information
df.describe()
```

```python
sns.countplot (x='species',data=df)
plt.title("Penguin Species Count")
plt.show()
```

```python
sns.countplot(x='island',data=df)
plt.title("Island-wise distribution")
plt.show()
```

```python
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=df, s=70)
plt.title("Flipper Length vs Body Mass")
plt.show()
```

```python
sns.pairplot(df,palette="pastel",hue="species")
plt.show()
```
